# 1. Physics Example

```python
import matplotlib.pyplot as plt # for making plots
import numpy as np # for easier dealing with arrays
```

```python
dat = np.loadtxt('calorimeter_data_v2.txt', skiprows=24)

dat = dat.reshape(-1,8,8)
dat = dat.transpose(0,2,1)

plt.imshow(dat[1])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
```

```python
import random

dat2 = dat

for i in range(dat2.shape[0]):
    if (random.random()>0.9 or np.sum(dat2[i])<-100):
        # if ( np.sum(dat2[i])>100 and np.sum(dat2[i])<350 or np.sum(dat2[i])>700 ):
        x,y = np.where(dat2[i] == np.amax(dat2[i]))
        x = x[0]
        y = y[0]
        # dat2[i,x,y] = 900 + random.uniform(-50,50)
        dat2[i] /= 100
        dat2[i,x,y] = random.gauss(800,60)

maxs = []
sums = []
for event in dat2:
    maxs.append(np.amax(event))
    sums.append(np.sum(event))
    
plt.hist(sums, bins=50)
plt.show()
```

```python
dat3 = dat2.transpose(0,2,1)

dat3 = dat3.reshape(-1,64)

dat3.shape

f = open('calorimeter_data_v3.txt', 'w')

for i in range(dat3.shape[0]):
    for j in range(dat3.shape[1]):
        f.write(f'{dat3[i,j]} ')
    f.write('\n')
    
f.close()
```

```python
dat4 = np.loadtxt('calorimeter_data_v3.txt', skiprows=24)

dat4 = dat4.reshape(-1,8,8)
dat4 = dat4.transpose(0,2,1)

maxs = []
sums = []
for event in dat4:
    maxs.append(np.amax(event))
    sums.append(np.sum(event))
    
plt.hist(sums, bins=40)
plt.show()
```
